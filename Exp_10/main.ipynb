## üêç Python Program: Data Visualization using Matplotlib & Seaborn

```python
# ======================================================
# Data Visualization using Matplotlib and Seaborn
# ======================================================

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set default Seaborn theme
sns.set(style="whitegrid")

# ------------------------------------------------------
# Load Sample Dataset
# ------------------------------------------------------
iris = sns.load_dataset("iris")
print(iris.head())

# ------------------------------------------------------
# 1Ô∏è‚É£ Line Plot (Matplotlib)
# ------------------------------------------------------
plt.figure(figsize=(6, 4))
plt.plot(iris.index, iris['sepal_length'], color='blue', label='Sepal Length')
plt.plot(iris.index, iris['petal_length'], color='red', label='Petal Length')
plt.title('Line Plot - Sepal vs Petal Length')
plt.xlabel('Index')
plt.ylabel('Length (cm)')
plt.legend()
plt.show()

# ------------------------------------------------------
# 2Ô∏è‚É£ Scatter Plot (Seaborn)
# ------------------------------------------------------
plt.figure(figsize=(6, 4))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris, palette='Set2')
plt.title('Scatter Plot - Sepal Length vs Petal Length')
plt.show()

# ------------------------------------------------------
# 3Ô∏è‚É£ Bar Plot (Seaborn)
# ------------------------------------------------------
plt.figure(figsize=(6, 4))
sns.barplot(x='species', y='sepal_width', data=iris, palette='pastel', ci=None)
plt.title('Bar Plot - Average Sepal Width per Species')
plt.show()

# ------------------------------------------------------
# 4Ô∏è‚É£ Histogram (Matplotlib)
# ------------------------------------------------------
plt.figure(figsize=(6, 4))
plt.hist(iris['petal_length'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram - Petal Length Distribution')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()

# ------------------------------------------------------
# 5Ô∏è‚É£ Box Plot (Seaborn)
# ------------------------------------------------------
plt.figure(figsize=(6, 4))
sns.boxplot(x='species', y='sepal_length', data=iris, palette='coolwarm')
plt.title('Box Plot - Sepal Length by Species')
plt.show()

# ------------------------------------------------------
# 6Ô∏è‚É£ Violin Plot (Seaborn)
# ------------------------------------------------------
plt.figure(figsize=(6, 4))
sns.violinplot(x='species', y='petal_width', data=iris, palette='Set3')
plt.title('Violin Plot - Petal Width by Species')
plt.show()

# ------------------------------------------------------
# 7Ô∏è‚É£ Heatmap (Seaborn)
# ------------------------------------------------------
plt.figure(figsize=(6, 4))
corr = iris.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Heatmap - Correlation Matrix')
plt.show()

# ------------------------------------------------------
# 8Ô∏è‚É£ Pair Plot (Seaborn)
# ------------------------------------------------------
sns.pairplot(iris, hue='species', palette='husl')
plt.suptitle('Pair Plot - Feature Relationships', y=1.02)
plt.show()

# ------------------------------------------------------
# 9Ô∏è‚É£ Count Plot (Seaborn)
# ------------------------------------------------------
plt.figure(figsize=(5, 4))
sns.countplot(x='species', data=iris, palette='Set1')
plt.title('Count Plot - Number of Samples per Species')
plt.show()

# ------------------------------------------------------
# üîü Pie Chart (Matplotlib)
# ------------------------------------------------------
species_count = iris['species'].value_counts()
plt.figure(figsize=(5, 5))
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Pie Chart - Species Distribution')
plt.show()
```

```python
# ================================================================
# EXPERIMENT 10: DATA VISUALIZATION USING MATPLOTLIB AND SEABORN
# Dataset: CAR DETAILS FROM CAR DEKHO.csv
# ================================================================

# -------------------------------
# Step 1: Import Required Libraries
# -------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

# Set the visualization style
sns.set(style="whitegrid")

# -------------------------------
# Step 2: Load the Dataset
# -------------------------------
df = pd.read_csv("Cars_details.csv")

# Display first few records
print("----- FIRST 5 ROWS OF DATASET -----")
print(df.head())

# Basic information about dataset
print("\n----- DATASET INFO -----")
print(df.info())

# Check summary statistics
print("\n----- SUMMARY STATISTICS -----")
print(df.describe())

# -------------------------------
# Step 3: Data Cleaning
# -------------------------------
print("\n----- MISSING VALUES CHECK -----")
print(df.isnull().sum())

# Drop rows with missing data (if any)
df = df.dropna()

# Create new column: Car Age
current_year = datetime.datetime.now().year
df['car_age'] = current_year - df['year']

# -------------------------------
# Step 4: Basic Exploration
# -------------------------------
print("\n----- UNIQUE VALUES -----")
print("Fuel Types:", df['fuel'].unique())
print("Seller Types:", df['seller_type'].unique())
print("Transmission Types:", df['transmission'].unique())
print("Owner Types:", df['owner'].unique())

# ================================================================
# Step 5: VISUALIZATION TECHNIQUES
# ================================================================

# 1Ô∏è‚É£ Histogram ‚Äî Distribution of Selling Price
plt.figure(figsize=(6,4))
plt.hist(df['selling_price'], bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Car Selling Price")
plt.xlabel("Selling Price (‚Çπ)")
plt.ylabel("Count")
plt.show()

# 2Ô∏è‚É£ Bar Plot ‚Äî Average Selling Price by Fuel Type
plt.figure(figsize=(6,4))
sns.barplot(x='fuel', y='selling_price', data=df, palette='Set2', ci=None)
plt.title("Average Selling Price by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Average Selling Price (‚Çπ)")
plt.show()

# 3Ô∏è‚É£ Count Plot ‚Äî Number of Cars by Transmission Type
plt.figure(figsize=(5,4))
sns.countplot(x='transmission', data=df, palette='coolwarm')
plt.title("Count of Cars by Transmission Type")
plt.xlabel("Transmission Type")
plt.ylabel("Count")
plt.show()

# 4Ô∏è‚É£ Scatter Plot ‚Äî Present Price vs Selling Price
plt.figure(figsize=(6,4))
sns.scatterplot(x='present_price', y='selling_price', hue='fuel', data=df, palette='Set1')
plt.title("Present Price vs Selling Price (by Fuel Type)")
plt.xlabel("Present Price (‚Çπ)")
plt.ylabel("Selling Price (‚Çπ)")
plt.show()

# 5Ô∏è‚É£ Box Plot ‚Äî Selling Price by Seller Type
plt.figure(figsize=(6,4))
sns.boxplot(x='seller_type', y='selling_price', data=df, palette='pastel')
plt.title("Box Plot - Seller Type vs Selling Price")
plt.xlabel("Seller Type")
plt.ylabel("Selling Price (‚Çπ)")
plt.show()

# 6Ô∏è‚É£ Violin Plot ‚Äî Car Age vs Fuel Type
plt.figure(figsize=(6,4))
sns.violinplot(x='fuel', y='car_age', data=df, palette='Set3')
plt.title("Violin Plot - Car Age Distribution by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Car Age (Years)")
plt.show()

# 7Ô∏è‚É£ Heatmap ‚Äî Correlation Between Numeric Columns
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Heatmap - Correlation Between Features")
plt.show()

# 8Ô∏è‚É£ Pair Plot ‚Äî Relationships Between Numeric Features
sns.pairplot(df[['year', 'selling_price', 'present_price', 'km_driven']], diag_kind='kde', palette='husl')
plt.suptitle("Pair Plot - Relationships Between Key Variables", y=1.02)
plt.show()

# 9Ô∏è‚É£ Pie Chart ‚Äî Distribution of Fuel Types
fuel_counts = df['fuel'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=120, colors=sns.color_palette('pastel'))
plt.title("Pie Chart - Fuel Type Distribution")
plt.show()

# üîü Line Plot ‚Äî Average Selling Price by Year
avg_price_by_year = df.groupby('year')['selling_price'].mean().sort_index()
plt.figure(figsize=(6,4))
plt.plot(avg_price_by_year.index, avg_price_by_year.values, marker='o', color='purple')
plt.title("Average Selling Price by Car Manufacturing Year")
plt.xlabel("Year")
plt.ylabel("Average Selling Price (‚Çπ)")
plt.grid(True)
plt.show()

```

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("Cars_details.csv")

# Display first few rows
print("Dataset Preview:")
print(df.head())

# Clean data: drop missing values
df = df.dropna()

# ------------------------------
# 1Ô∏è‚É£ Basic Info
# ------------------------------
print("\nDataset Info:")
print(df.info())

# ------------------------------
# 2Ô∏è‚É£ Histogram - Distribution of Car Prices
# ------------------------------
plt.figure(figsize=(8,5))
plt.hist(df['selling_price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Selling Price')
plt.xlabel('Selling Price (in ‚Çπ)')
plt.ylabel('Frequency')
plt.show()

# ------------------------------
# 3Ô∏è‚É£ Boxplot - Outlier Detection in Price
# ------------------------------
plt.figure(figsize=(6,5))
sns.boxplot(x=df['selling_price'], color='orange')
plt.title('Boxplot of Selling Price')
plt.show()

# ------------------------------
# 4Ô∏è‚É£ Countplot - Fuel Type Distribution
# ------------------------------
plt.figure(figsize=(6,5))
sns.countplot(x='fuel', data=df, palette='coolwarm')
plt.title('Count of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.show()

# ------------------------------
# 5Ô∏è‚É£ Barplot - Average Price by Fuel Type
# ------------------------------
plt.figure(figsize=(6,5))
sns.barplot(x='fuel', y='selling_price', data=df, estimator='mean', palette='Set2')
plt.title('Average Selling Price by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Average Price (‚Çπ)')
plt.show()

# ------------------------------
# 6Ô∏è‚É£ Scatter Plot - Year vs Selling Price
# ------------------------------
plt.figure(figsize=(7,5))
plt.scatter(df['year'], df['selling_price'], color='green', alpha=0.5)
plt.title('Selling Price vs Year')
plt.xlabel('Year')
plt.ylabel('Selling Price (‚Çπ)')
plt.show()

# ------------------------------
# 7Ô∏è‚É£ Pairplot - Relationships between Features
# ------------------------------
sns.pairplot(df[['year', 'selling_price', 'km_driven']], diag_kind='kde')
plt.suptitle("Pairplot of Numerical Features", y=1.02)
plt.show()

# ------------------------------
# 8Ô∏è‚É£ Heatmap - Correlation between Numerical Columns
# ------------------------------
plt.figure(figsize=(6,5))
corr = df[['year', 'selling_price', 'km_driven']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# ------------------------------
# 9Ô∏è‚É£ Pie Chart - Seller Type Distribution
# ------------------------------
plt.figure(figsize=(5,5))
seller_counts = df['seller_type'].value_counts()
plt.pie(seller_counts, labels=seller_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set3'))
plt.title('Seller Type Distribution')
plt.show()
```
