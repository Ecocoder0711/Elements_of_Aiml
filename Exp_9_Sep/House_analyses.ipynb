```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv("House_price_prediction.csv")

# Show first 5 rows
print("First 5 records:\n", data.head())

# Show basic info
print("\nDataset Info:")
print(data.info())

# Show summary statistics
print("\nSummary Statistics:")
print(data.describe())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Visualizations

# Price distribution
plt.figure(figsize=(8,5))
sns.histplot(data['price'], kde=True, bins=30)
plt.title("Price Distribution")
plt.show()

# Living Area vs Price (using sqft_living instead of area)
plt.figure(figsize=(8,5))
sns.scatterplot(x="sqft_living", y="price", hue="floors", data=data, palette="viridis")
plt.title("Living Area vs Price (colored by Floors)")
plt.show()

# Boxplot: Condition vs Price (using condition instead of furnishingstatus)
plt.figure(figsize=(8,5))
sns.boxplot(x="condition", y="price", data=data)
plt.title("Price vs Condition")
plt.show()

# Alternative: Waterfront vs Price
plt.figure(figsize=(8,5))
sns.boxplot(x="waterfront", y="price", data=data)
plt.title("Price vs Waterfront Property")
plt.show()

# Correlation heatmap
plt.figure(figsize=(10,8))
# Select only numeric columns for correlation
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
sns.heatmap(data[numeric_cols].corr(), annot=True, cmap="coolwarm", linewidths=0.5, fmt='.2f')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
