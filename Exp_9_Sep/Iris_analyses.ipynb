```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Load dataset
data = pd.read_csv("IRIS_flower.csv")

# 1. Swarm Plot (each data point shown by species)
plt.figure(figsize=(10,6))
sns.swarmplot(x="species", y="sepal_width", data=data, palette="Set2")
plt.title("Swarm Plot of Sepal Width by Species")
plt.show()

# 2. Strip Plot (similar to swarm but simpler)
plt.figure(figsize=(10,6))
sns.stripplot(x="species", y="petal_width", data=data, jitter=True, palette="Set1")
plt.title("Strip Plot of Petal Width by Species")
plt.show()

# 3. KDE Plot (distribution curves)
plt.figure(figsize=(10,6))
sns.kdeplot(data=data, x="sepal_length", hue="species", fill=True)
plt.title("KDE Distribution of Sepal Length")
plt.show()

# 4. Count Plot (counts of species)
plt.figure(figsize=(6,4))
sns.countplot(x="species", data=data, palette="pastel")
plt.title("Count of Each Iris Species")
plt.show()

# 5. 3D Scatter Plot (Sepal Length, Sepal Width, Petal Length)
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")
species_list = data["species"].unique()
colors = ["red", "green", "blue"]

for species, color in zip(species_list, colors):
    subset = data[data["species"] == species]
    ax.scatter(subset["sepal_length"], subset["sepal_width"], subset["petal_length"], 
               label=species, color=color, s=50)

ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
plt.title("3D Scatter Plot of Iris Features")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Rectangle
import warnings
warnings.filterwarnings('ignore')

# Load dataset
data = pd.read_csv("IRIS_flower.csv")

# Set the style for better visuals
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# 1. ADVANCED PAIRPLOT WITH CUSTOM STYLING
plt.figure(figsize=(14, 10))
g = sns.pairplot(data, hue="species", diag_kind="kde", 
                 plot_kws={"alpha": 0.7, "s": 60, "edgecolor": "k", "linewidth": 0.5},
                 diag_kws={"alpha": 0.7, "linewidth": 2})
g.fig.suptitle("Comprehensive Iris Dataset Analysis - Pairplot", fontsize=16, y=1.02)
plt.show()

# 2. MULTI-SUBPLOT VIOLIN PLOTS
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle("Violin Plots: Distribution Analysis by Species", fontsize=18, fontweight='bold')

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
titles = ['Sepal Length Distribution', 'Sepal Width Distribution', 
          'Petal Length Distribution', 'Petal Width Distribution']

for i, (feature, title) in enumerate(zip(features, titles)):
    row, col = i // 2, i % 2
    sns.violinplot(data=data, x="species", y=feature, ax=axes[row, col], 
                   inner="quart", palette="muted")
    axes[row, col].set_title(title, fontsize=14, fontweight='bold')
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 3. CORRELATION HEATMAP WITH ANNOTATIONS
plt.figure(figsize=(10, 8))
correlation_matrix = data.select_dtypes(include=[np.number]).corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

sns.heatmap(correlation_matrix, mask=mask, annot=True, cmap='RdYlBu_r', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8},
            fmt='.3f', annot_kws={'size': 12, 'weight': 'bold'})
plt.title("Feature Correlation Matrix\n(Lower Triangle)", fontsize=16, fontweight='bold', pad=20)
plt.show()

# 4. RADAR CHART (SPIDER PLOT)
from math import pi

categories = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
species_means = data.groupby('species')[categories].mean()

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))

angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
species_list = species_means.index

for i, species in enumerate(species_list):
    values = species_means.loc[species].values.flatten().tolist()
    values += values[:1]
    
    ax.plot(angles, values, 'o-', linewidth=3, label=species, color=colors[i])
    ax.fill(angles, values, alpha=0.25, color=colors[i])

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, fontsize=12)
ax.set_ylim(0, max(species_means.values.flatten()) * 1.1)
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1.0))
plt.title("Species Characteristics Radar Chart", size=16, fontweight='bold', pad=30)
plt.show()

# 5. ADVANCED 3D SCATTER WITH GRADIENT COLORS
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Check actual species names in your dataset
print("Actual species names in dataset:", data['species'].unique())

# Create dynamic color mapping based on actual species names
unique_species = data['species'].unique()
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
colors_map = {species: colors[i] for i, species in enumerate(unique_species)}

for species in data['species'].unique():
    subset = data[data['species'] == species]
    ax.scatter(subset['sepal_length'], subset['sepal_width'], subset['petal_length'],
               c=colors_map[species], label=species, s=60, alpha=0.8, edgecolors='black', linewidth=0.5)

ax.set_xlabel('Sepal Length (cm)', fontsize=12, fontweight='bold')
ax.set_ylabel('Sepal Width (cm)', fontsize=12, fontweight='bold')
ax.set_zlabel('Petal Length (cm)', fontsize=12, fontweight='bold')
ax.set_title('3D Feature Space Visualization', fontsize=16, fontweight='bold', pad=20)
ax.legend(fontsize=12)

# Add grid
ax.grid(True, alpha=0.3)
plt.show()

# 6. BOX PLOT WITH SWARM OVERLAY
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Box Plots with Data Points Overlay", fontsize=18, fontweight='bold')

for i, feature in enumerate(features):
    row, col = i // 2, i % 2
    
    # Box plot
    sns.boxplot(data=data, x="species", y=feature, ax=axes[row, col], 
                palette="Set2", width=0.6)
    
    # Swarm plot overlay
    sns.swarmplot(data=data, x="species", y=feature, ax=axes[row, col], 
                  color="black", alpha=0.6, size=4)
    
    axes[row, col].set_title(f'{feature.replace("_", " ").title()} Analysis', 
                            fontsize=14, fontweight='bold')
    axes[row, col].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 7. RIDGE PLOT (MULTIPLE DENSITY CURVES)
from scipy import stats

fig, ax = plt.subplots(figsize=(12, 8))

y_pos = 0
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
feature = 'petal_length'

for i, species in enumerate(data['species'].unique()):
    subset = data[data['species'] == species][feature]
    
    # Create density curve
    density = stats.gaussian_kde(subset)
    xs = np.linspace(subset.min(), subset.max(), 200)
    density_values = density(xs)
    
    # Normalize and offset
    density_values = density_values / density_values.max() * 0.8
    
    ax.fill_between(xs, y_pos, y_pos + density_values, 
                    alpha=0.7, color=colors[i], label=species)
    ax.plot(xs, y_pos + density_values, color='black', linewidth=1)
    
    # Add species label
    ax.text(xs[0] - 0.3, y_pos + 0.4, species, fontsize=12, fontweight='bold')
    
    y_pos += 1

ax.set_xlabel(f'{feature.replace("_", " ").title()} (cm)', fontsize=12, fontweight='bold')
ax.set_ylabel('Species', fontsize=12, fontweight='bold')
ax.set_title('Ridge Plot: Petal Length Distribution by Species', fontsize=16, fontweight='bold')
ax.set_yticks([])
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 8. STATISTICAL SUMMARY TABLE
summary_stats = data.groupby('species').agg({
    'sepal_length': ['mean', 'std', 'min', 'max'],
    'sepal_width': ['mean', 'std', 'min', 'max'],
    'petal_length': ['mean', 'std', 'min', 'max'],
    'petal_width': ['mean', 'std', 'min', 'max']
}).round(2)

print("\n" + "="*80)
print("COMPREHENSIVE STATISTICAL SUMMARY BY SPECIES")
print("="*80)
print(summary_stats)
print("="*80)

# 9. FEATURE IMPORTANCE VISUALIZATION
feature_ranges = []
for feature in features:
    species_means = data.groupby('species')[feature].mean()
    feature_range = species_means.max() - species_means.min()
    feature_ranges.append(feature_range)

plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(features)), feature_ranges, 
               color=['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A'])

plt.xlabel('Features', fontsize=12, fontweight='bold')
plt.ylabel('Range of Species Means', fontsize=12, fontweight='bold')
plt.title('Feature Discriminative Power\n(Higher = Better for Species Classification)', 
          fontsize=14, fontweight='bold')
plt.xticks(range(len(features)), [f.replace('_', ' ').title() for f in features])

# Add value labels on bars
for bar, value in zip(bars, feature_ranges):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{value:.2f}', ha='center', va='bottom', fontweight='bold')

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n Analysis Complete! These visualizations showcase:")
print(" Advanced statistical distributions")
print(" Multi-dimensional relationships") 
print(" Professional styling and annotations")
print(" Comprehensive feature analysis")
print(" Interactive 3D visualizations")
```
