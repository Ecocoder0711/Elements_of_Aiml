# Analysing datasets using only pandas
we have have two datasets and we have to compare them with only using pandas.
the datasets we have are:
1. House_price_prediction.csv
2. IRIS_flower.csv 

```python
import pandas as pd

df1 = pd.read_csv('IRIS_flower.csv')
print(df1)
df2 = pd.read_csv('House_price_prediction.csv')
print(df2)
```

Lets's try to first analyses both data usnig `info()`

```python
# Dataset-1 : IRIS_flower
print('Iris_flower_dataset')
print(df1.info())
# Dataset-2 : House_price_prediction
print('House_price_prediction')
print(df2.info())
```

we can also take help to analyses data with `corr()`.. we can find correlation between numerical columns.

```python
print("Correlation of dataset 1\n")
print(df1.corr(numeric_only = True))
print()
print('Correlation of dataset 2\n')
print(df2.corr(numeric_only = True))
```
